```python
import numpy as np

class Neuron:
    def __init__(self, num_inputs):
        self.weights = np.array([0.1, 0.2])  # Correct initialization of weights
        self.bias = 0

    def activation(self, z):
        return 1 / (1 + np.exp(0.5 * (-z)))  # Sigmoid activation function with scaling

    def forward(self, inputs):
        z = np.dot(inputs, self.weights) + self.bias
        output = self.activation(z)
        return output

class Layer:
    def __init__(self, num_inputs):
        self.neuron = Neuron(num_inputs)

    def forward(self, inputs):
        return self.neuron.forward(inputs)

class NeuralNetwork:
    def __init__(self, input_size):
        self.input_layer = Layer(input_size)
        self.hidden_layer = Layer(1)  # Single neuron in hidden layer
        self.output_layer = Layer(1)  # Single neuron in output layer

    def forward(self, inputs):
        out1 = self.input_layer.forward(inputs)
        out2 = self.hidden_layer.forward([out1])
        out3 = self.output_layer.forward([out2])
        return out3

# Initialize neural network with 2 inputs
nn = NeuralNetwork(input_size=2)

# Example input data with 2 inputs
input_data = np.array([1, 2])

# Perform a forward pass
output = nn.forward(input_data)

# Print the output of the network
print("Output of the network:", output)
```

```python
import numpy as np

class Neuron:
    def __init__(self, num_inputs):
        
        self.weights = np.array([0.1, 0.2])
        self.bias = 0               

    def activation(self, z):
        
        return 1 / (1 + np.exp(0.5 * (-z)))

    def forward(self, inputs):
        
        z = np.dot(inputs, self.weights) + self.bias
        output = self.activation(z)
        return output

class Layer:
    def __init__(self, num_neurons, num_inputs_per_neuron):
        
        self.neurons = [Neuron(num_inputs_per_neuron) for _ in range(num_neurons)]

    def forward(self, inputs):
        
        outputs = [neuron.forward(inputs) for neuron in self.neurons]
        return outputs

class NeuralNetwork:
    def __init__(self):
        
        self.input_layer = Layer(2, 2)  
        self.hidden_layer = Layer(2, 2)  
        self.output_layer = Layer(1, 2)  

    def forward(self, inputs):
        
        hidden_outputs = self.input_layer.forward(inputs)
        
        
        hidden_layer_outputs = self.hidden_layer.forward(hidden_outputs)
        
        
        final_output = self.output_layer.forward(hidden_layer_outputs)
        
        return final_output


nn = NeuralNetwork()


input_data = np.array([1, 2])  


output = nn.forward(input_data)


print("Output of the network:", output)
```
